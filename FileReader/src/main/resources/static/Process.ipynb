```python
import pandas as pd
```

## xlsx操作

```python
data = pd.read_excel("1.xlsx", sheet_name=None)
data
```

```python
writer = pd.ExcelWriter("2.xlsx")
for df in data:
    data[df] = data[df].head(20)
    data[df].to_excel(writer, sheet_name=df, index=False)
writer.save()
```
